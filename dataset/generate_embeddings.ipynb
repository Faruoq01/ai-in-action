```python
import vertexai
from vertexai.language_models import TextEmbeddingModel
import os
from dotenv import load_dotenv
import certifi

load_dotenv()
```

```python
def generate_embeddings(query):
    vertexai.init(project="gen-lang-client-0982532504", location="us-central1")
    model = TextEmbeddingModel.from_pretrained("text-embedding-004")
    embeddings = model.get_embeddings([query])
    return embeddings[0].values
```

```python
text = "The patient's son has adult-onset Still's disease (AOSD), previously known as systemic-onset juvenile idiopathic arthritis (SoJIA) when under 16. His symptoms include a rash that resolves with fever, muscle aches, joint pain, and enlarged spleen and liver, with elevated CRP, ESR, ASO, and ferritin levels. Liver enzymes are also elevated due to swelling. He is currently on high-dose prednisone and biologic injections and was hospitalized for three weeks. The clinician emphasizes the need for aggressive early treatment for better chances of remission and suggests sharing information about SoJIA symptoms with the patient's mother for better understanding."
embedding = generate_embeddings(text)
embedding
```
